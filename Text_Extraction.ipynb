```python
#TEXT SEGMENTS EXTRACTION
import os
import re
from nltk.tokenize import word_tokenize
import pandas as pd
import logging
import csv

Max_Tokens = 8192
Overlap = 500

# Set up logging configuration
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Function to clean text by removing specific patterns
def clean_text(text):
    # Define patterns of useless information (example: remove all numbers)
    patterns = [
        r'\((X|\?|☐)\)\s*\w+',  # Pattern to match checkboxes
        r'[\|\-\_]{2,}'  # Pattern to match repeated '|', '-', '_'
        # Add more patterns here
    ]
    for pattern in patterns:
        text = re.sub(pattern, '', text)  # Replace with a single space
    return text.strip()

# Generator function to segment text into overlapping chunks
def segment_text(text, max_tokens=Max_Tokens, overlap=Overlap):
    words = word_tokenize(text)
    start = 0
    while True:
        end = min(start + max_tokens, len(words))
        segment = ' '.join(words[start:end])
        yield segment
        if end == len(words):
            break
        start = end - overlap

# Function to process a single file
def process_file(file_path, filename):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        text = file.read()
    
    # Extract checkboxes before cleaning
    checkbox_pattern = re.compile(r'\((X|\?|☐)\)\s*(\w+)')
    checkbox_matches = checkbox_pattern.findall(text)
    checked_labels = ', '.join([label for state, label in checkbox_matches if state == 'X'])
    
    # Clean the text
    text = clean_text(text)
    
    # Estimate the number of segments
    words = word_tokenize(text)
    estimated_segments = max(1, (len(words) - Overlap) // (Max_Tokens - Overlap) + 1)
    
    logging.info(f"Estimated {estimated_segments} segments for {filename}")
    
    # Segment the text and process each segment
    segment_counter = 0
    for segment in segment_text(text):
        segment_counter += 1
        if estimated_segments > 1 and segment_counter % 1 == 0:
            logging.info(f"Processed segment {segment_counter} of {estimated_segments} for file: {filename}")
        
        # Append the segment's data to the CSV file
        with open(output_csv, mode='a', encoding='utf-8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow([filename, checked_labels, segment])

# Specify the directory containing the .txt files
directory = 'RegulatoryBulletin/Files'

# Specify the output CSV file
output_csv = 'segments_extraction.csv'

# Define the columns for the CSV file
columns = ['Filename', 'Checked', 'Segment']

# Remove the output CSV file if it already exists
if os.path.exists(output_csv):
    logging.info(f"Removing existing output CSV file: {output_csv}")
    os.remove(output_csv)

# Write headers to the CSV file
with open(output_csv, mode='w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(columns)

# Iterate over each file in the directory and process them
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        logging.info(f"Starting file: {filename}")
        process_file(file_path, filename)
        logging.info(f"Finished file: {filename}")

# Print completion message
logging.info('Data processing complete.')
```

```python
# KEYWORD EXTRACTION 
import os
import re
import pandas as pd
import logging
import csv
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Set up logging configuration
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Function to clean text by removing specific patterns
def clean_text(text):
    patterns = [
        r'\((X|\?|☐)\)\s*\w+',  # Pattern to match checkboxes
        r'[\|\-\_]{2,}'  # Pattern to match repeated '|', '-', '_'
        # Add more patterns here
    ]
    for pattern in patterns:
        text = re.sub(pattern, '', text)  # Replace with a single space
    return text.strip()

# Function to extract keywords from a single file
def extract_keywords(text, top_n=250):
    try:
        # Tokenize the text
        words = word_tokenize(text)
        # Remove stopwords
        words = [word for word in words if word.lower() not in stopwords.words('english')]
        # Join the words back into a string
        cleaned_text = ' '.join(words)
        
        # Check if cleaned_text is empty
        if not cleaned_text.strip():
            raise ValueError("The cleaned text is empty after removing stopwords.")
        
        # Initialize TF-IDF Vectorizer
        vectorizer = TfidfVectorizer(stop_words='english')
        tfidf_matrix = vectorizer.fit_transform([cleaned_text])
        
        # Get feature names
        feature_names = vectorizer.get_feature_names_out()
        # Get the scores for each feature
        tfidf_scores = tfidf_matrix[0].toarray().flatten()
        # Create a mapping of scores to words
        word_scores = zip(feature_names, tfidf_scores)
        # Sort the words by their score
        sorted_word_scores = sorted(word_scores, key=lambda x: x[1], reverse=True)
        
        # Extract the top n keywords
        keywords = [word for word, score in sorted_word_scores[:top_n]]
        return ', '.join(keywords)
    except ValueError as e:
        # Handle the case where the text is empty or only contains stopwords
        return "No keywords could be extracted."

# Function to process a single file
def process_file(file_path, filename):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        text = file.read()
    
    # Extract checkboxes before cleaning
    checkbox_pattern = re.compile(r'\((X|\?|☐)\)\s*(\w+)')
    checkbox_matches = checkbox_pattern.findall(text)
    checked_labels = ', '.join([label for state, label in checkbox_matches if state == 'X'])
    
    # Clean the text
    text = clean_text(text)
    
    # Extract keywords
    keywords = extract_keywords(text)
    
    logging.info(f"Extracted keywords for {filename}")
    
    # Append the file's data to the CSV file
    with open(output_csv, mode='a', encoding='utf-8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([filename, checked_labels, keywords])

# Specify the directory containing the .txt files
directory = 'RegulatoryBulletin/Files'

# Specify the output CSV file
output_csv = 'keyword_extraction.csv'

# Define the columns for the CSV file
columns = ['Filename', 'Checked', 'Keywords']

# Remove the output CSV file if it already exists
if os.path.exists(output_csv):
    logging.info(f"Removing existing output CSV file: {output_csv}")
    os.remove(output_csv)

# Write headers to the CSV file
with open(output_csv, mode='w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(columns)

# Iterate over each file in the directory and process them
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        logging.info(f"Starting file: {filename}")
        process_file(file_path, filename)
        logging.info(f"Finished file: {filename}")

# Print completion message
logging.info('Keyword extraction complete.')
```

## Convert Keyword Extraction CSV Data to DPO Format

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('keyword_extraction.csv')

# Combine 'Filename' and 'Checked' columns into a new 'question' column
df['question'] = df['Filename'].astype(str) + ',' + df['Checked'].astype(str)

# Rename the 'Keywords' column to 'chosen'
df.rename(columns={'Keywords': 'chosen'}, inplace=True)

# Drop the original 'Filename' and 'Checked' columns as they are no longer needed
df.drop(columns=['Filename', 'Checked'], inplace=True)

# Vectorize the 'question' column
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['question'])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=0).fit(X)  # Adjust n_clusters as needed
df['cluster'] = kmeans.labels_

# Function to find the row with the most different items but at least 2 repeated items within a cluster
def find_rejected(row, cluster_df):
    question_items = set(row['question'].split(','))
    max_diff = 0
    rejected_value = None
    for _, other_row in cluster_df.iterrows():
        if other_row.name == row.name:
            continue
        other_items = set(other_row['question'].split(','))
        common_items = question_items & other_items
        different_items = question_items ^ other_items
        if len(common_items) >= 2 and len(different_items) > max_diff:
            max_diff = len(different_items)
            rejected_value = other_row['chosen']
    return rejected_value

# Apply the function to each row within its cluster
df['rejected'] = df.apply(lambda row: find_rejected(row, df[df['cluster'] == row['cluster']]), axis=1)

# Drop the 'cluster' column as it is no longer needed
df.drop(columns=['cluster'], inplace=True)

# Save the transformed DataFrame to a new CSV file
df.to_csv('gpce_dpo.csv', index=False)
```
